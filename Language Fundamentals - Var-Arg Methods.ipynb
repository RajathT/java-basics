### Varaible Number of Argument Methods
Until 1.4 V we cannot declare with variable number of arguments, if there is a change of number of arguments, we have to go for a new method. To overcome this problem, var-arg method 1.5V arrcoding to this, we can decalre a method which can take vaiable number of methods.
We can declare a var - arg methods as follows.
m1 (int... x)
We can call any number of values, even 0.
Internally var-arg method will be converted to one dimensional array,hence we can differentiate values but using index.

### Case 1:
~~~java
m1 (int ...x);
m2 (int... x);
m3 (int. ..x);//Invalid
m4 (int .x..);//Invalid
~~~

### Case 2:
we can mix var arg paramenter with normal parameter
~~~java
m1 (int x, int ...x);
~~~

### Case 3:
~~~java
m2 (int ...x, String s);//Invalid
~~~

### Case 4:
We can take only one var arg parameter.

~~~java
m2 (int ...x, String ...s);//Invalid
~~~

### Case 5:
Inside a class we cannot declare var arg method and corresponding 1D method simultanously.

~~~java
//Invalid
class Test{
    public static void m1(int[]x){
    }
    public static void m1(int...x){
    }
}
//C.E: Cannot declare both m1(int[]) and m1(int..) in same class Test
~~~

### Case 6:
~~~java
class Test{

    public static void m1(int x){
        //General
    }
    public static void m1(int...x){
    }
    public static void main(String []args){
        m1();//... method
        m1(10,20);//...
        m1(10);//General
    }
}
~~~

Least preference is given to var-arg, new features are given least priority. If no other method matched then only var-arg method matches, this is like default in switch.

### Equivalence between var arg parameter and 1D array

### Case 1:
Where ever 1D array present we can replace with var-arg parameter. m1(int []x) => m1(int... x)
main(String[] args) => main(String... args)

### Case 2:
Wherever var-arg parameter is present we cannot replace with 1D array
m1(int...x) => m1(int[]x)//Invalid m1() will not work
m1(10,20,30) - will not work
m1(new int[]{10,20,20}) - this will work

### Case 3:
Note m1(int...x) => we can call this method by passing a group of int values and it will become 1D array
m1(int[]...x) we can call this method by passing a group on 1D arrays, and x will become 2D int array.

~~~java
class Test{
    public static void main(String []args){
        int []a = {10,20,30};
        int []b = {40,50,60};
        
        m1(a,b);
    }
    public static void m1(int []...x){
        for (int []x1: x)
            System.out.println(x1[0]);//Output: 10,40
    }
}
~~~
