### Literals
Any constant value that can be assigned to a variable is called Literals.

int - keyword
x = - identifier
10 - Literal

#### Integral literals:

For intergrals DT's like byte, short, int long, we can specify literal values in the following ways.

Decimal from base10 allowed digits are 0-9
Octal for base8 allowed digits are 0-7, and should be prefixed with 0, for. e,g 0777 is valid
Hexadecimal for base16 allowed digits are 0-9, a-f we can use both lower case or upper case and should be prefixed by 0X.

These are the only ways to specify integral values.

JVM will print these values is decimal form.

By default every integral value is of int type, but we can specify as long type by l or L.

int x = 10;
long l = 10L;
int x = 10L; // P.L.P
long l = 10;

No direct way to specify byte, short, if they fall in the range then they satisfy else, C.E: P.L.P

#### Floating point literals:

By default, every FPL is of double type, hence we cannot assign directly to the float variable.

float f = 12.34; P.L.P found double, req float

We can assign decimal, octal, hexadecimal to floating point literals.

#### Boolean literals
We can assign true/false. For other type of assignments we get C.E: Incompatible types or var not found.

#### char literals

We can assign any value between 0-65535 to the char which are of intergral type.

byte b = 123;
char ch = b;
System.out.println(ch);//Valid O/p: {

char ch = 0xface;
or anything of integral number which fall under the range.

We can also represent char literals in unicode format also.
'\uxxx'. 4 digit hexadecimal number.

char ch = '\u0061';
System.out.println(ch);//a

Every escape character is a valid char literal:

char ch = '\n';

Escape character are:

\n - newline
\t - horizontal tab
\r - carriage return
\b - back space
\f - form feed
\' - single quote
\" - double quote
\\ - backslash

#### string literal

Any sequence of characters are treated as string literals.

String s = "rajath";

#### Binary literals

In 1.7v for integral data types beside from dec, oct, hexa they have added binary.

int x = 0b1111;
System.out.println(x);//15

#### Usage of _ in java

To distinguish the digits and improve readability. in java we can use. 1_2_2 which will be readed as 122.

We have to use the _ in between and not on the start or end.
