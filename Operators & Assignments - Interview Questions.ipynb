### new vs newInstance

We can use new operator to create an object if we know the class name at the beginning.

Test t = new Test();
Student s = new Student();
Customer c = new Customer();

newInstance is a method present in class Class, we can use newInstance method if we don't know the class name at the beginning.

~~~java


class Student{}
class Customer{}

class Test{
    public static void main(String[]args){
        System.out.println(args[0]);
        Object o = Class.forName(args[0]).newInstance();
    }
}
~~~

In the case of new operator based on our req we can invoke any constructor

Test t = new Test();
Test t = new Test(10);

But newInstance method calls no arg constructor, so to use newInstance, compulsory the class should contain no arg constructor. We get Runtime exception: InstantiationException.

~~~java
Test t = new Test();
~~~

While using new operator at runtime if the corresponding .class file is not available then we will get run time exception: NoClassDefFoundError: Test

at runtime if Test.class not available, then we will get runtime exception saying: Test
~~~java
Object o = Class.forName(args[0]).newInstance();

//java Test Test123
//R.E: ClassNotFoundException: Test123
~~~
While using newInstance method at runtime if the corresponding class file not available then we will get R.E: ClassNotFoundException.

### classNotFoundException and NoClassDefFoundError

For har coded class names, at runtime if the corresponding.class file is not available, then we get R.E: NoClassDefFoundError, which is unchecked.

Test t = new Test();
At Runtime, if Test.class is not available we get RE: NoClassDefFoundError: Test

For dynamically provided class names, if the corresponding .class file is not available, then we will get runtime exception the we get : ClassNotFoundException which is a checked exception.

### instanceof vs isInstance()

instanceof is an operator in Java, we can use instanceof to check whether the given obj is of a particular type or not and we know the type at the beginning.
~~~java
Thread t = new Thread();

System.out.println(t instanceof Runnable);
~~~
isInstance() is present in java.lang.Class we can use isInstance() to check whether of a particular type or not and we don't know the type at the beginning and it's available at the runtime.
~~~java
class Test{
    public static void main(String []args){
        Thread t = new Thread();
        System.out.println(Class.forName(args[0]).isInstance());
    }
}
~~~

java Test java.lang.Runnable//true

isInstance() is a method equivalent of isinstance operator.
