## Division 2
Based on position of declartion and behaviour, all variables are divided into 3 types.
1. Instance Variables: 
    * If the value of the variable is varied from obj to obj, such type of variables are called instance variables.
    * For every obj a seperate copy of instance variable will be created.
    * They should be declared within the class directyl but outside of any method or constructor.
    * Instance var will be created at the time of obj creation and destroyed at the type of obj destruction, hence the scope of instance variable is same as obj.
    * instace variables will be stored in the heap mem as a part of obj.
    * we cant access instance variables directly from static area but we can access by using obj reference.
    * but we can access instance from instance area
~~~java
class Test{
    int x = 10;
    public static void main(String [] args){
        System.out.println(x);
        /* Compiletime Error: 
            Non static variable x cannot be referenced from a static context
        */
        Test t = new Test();
        System.out.println(t.x);//Valid
    }
    public void m(){
        System.out.println(x);//Valid
    }
}
~~~
    * For instance variables the JVM will provide default values, do not need to perform initialization explicitly
    * Instnce variables also known as obj level variables or attributes
2. Static Variable:
    * If the value of a var is not varied from obj to obj then it is not recommended to declare as a instance variable, we have to declare such variables at the class level using the static modifier.
    * In the case of instance variables for every obj a seperate copy will be created but in the case of static variables a single copy will be created at class level and shared by every obj of the class
    * Static variables should be declared within the class directly but outside of any block or constructor.
    * Static variables will be created at the time of class loading and destroyed at the time of class unloading, hence the scope of static variable is same of the .class file.
    * How does a Java execute the class file:
        1. Start JVM
        2. Create and start main Thread
        3. locate Test.class file
        4. load Test.class - Static variable creation
        5. Execute the main() method
        6. Unload Test.class - Static variable destruction
        7. Terminate main Thread
        8. Shutdown JVM
        
    * static variables will be stored in method area
    * we can access static variables by obj ref or class name, but rcommended to use class name. Within the same class it is not req to use clss name, we can access directly.
~~~java
class Test{
    static int x = 10;
    public static void main(String []args){
        System.out.println(x);//Valid
        
        Test t = new Test();
        System.out.println(t.x);
        System.out.println(Test.x);
    }
}
~~~
    * we can access static variables from both instance and static areas.
    * For static variables JVM will provide default values and we are not req to perform initialization explicitly.
~~~java
class Test{
    static int x;
    static double d;
    static boolean b;
    static String s;
    
    public static void main(String []args){
        System.out.println(x);//0
        System.out.println(d);//0.0
        System.out.println(b);//false
        System.out.println(s);//null
    }
}
~~~  
~~~java
class Test{
    static int x = 10;
    int y = 20;

    public static void main(String []args){
        Test t1 = new Test();
        t1.x = 888;
        t1.y = 999;
        Test t2 = new Test();
        System.out.println(t2.x + t2.y);//888 20    
    }
}
~~~
    * Static variables are also known as Fields.
3. Local Variables:
    * Sometimes to meet temp req of the programmer, we can declare vars inside a method or block or constructor, such type of variables are called local variables or temp variables or automatic variables. 
    * Local variables will be stored in the stack memory.
    * Local variables will be created while executing the block in which we declared it, once block execution completes automatically local variables will be destroyed, hence the scope of local var is the block in which we declared it.

~~~java
class Test{
    
    int i = 0;

    public static void main(String []args){
         for (int j=0; j<2 ; j++){
             i = i+j;
         }
        System.out.println(i+j);//Compiler will give left and right
        /*
            CE : Cannot find symbol:
            variable j
            locatoin: class test
        */
    }
}
~~~

    * For local variables JVM won't provide default values compulsory to provide initialozation explicitly before using the variable. i.e if we care not using then it is not req to perform initialization.
    
~~~java
class Test{

    public static void main(String []args){
        int x;
~~~

    * It is not recommeneded to perform initialization for local variables inside logical blocks because there is guarntee of these blocks always at runtime
    * It is highly recommended to perform initialization for local variablse at the time of initialization with default values.
    * The only applicable modifier for local variables is final, by mistake if we are trying to apply any other modifer then we will get compile time error.
    * Only modifier of local var is final, cannot assign public, protected etc. C.E: Illegal start of expression.
    * local var access modifier is fixed, by default it is default, but this rule is applicable only for instance and static variables, but not for local variables.
    

## Conclusions
1. For instance and static variables JVM will provide default values, do not require initialozation explicitly but for local variables, JVM doesnot provide initilozation, we should perform initialization before use.
2. Instance and static variables can be accessd by multiple threads simultaneouly, hence not thread safe, but in the case of local variables for every thread a seperate copy will be created, hence they are thread safe.

|Type of variable|Thread Safe|
|-|-|
|instance variable|No|
|static variable|No|
|local variable|Yes|
