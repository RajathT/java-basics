### Increment and Decrement operators

|Expression|Initial value of x| value of y| final value of x|
|-|-|-|-|
|y = x++|10|11|11|
|y = ++x|10|10|11|
|y = x--|10|10|9|
|y = --x|10|9|9|

#### Case 1:
We can apply increment and decrement operators only for variables but not for constant values. If we are trying to apply for constant calues, we get compile time error.

int y = ++x;//Valid
int y = ++10;//Invalid
C.E: Unexpected type, found value required variable.

#### Case 2:
Nesting of In/Decrement not allowed

int x = ++(++y);//Invalid
C.E: Unexpected type, found value required variable.

#### Case 3:
For final variables we can't apply In/De operators

final int x = 10;
x++;//Invalid
C.E: cannot assign value to final variable

#### Case 4:
We can apply In/De operators for every primitive type except boolean
~~~java
int y = 10;
y++;

double y = 10.5;
y++;//Valid = 11.5

char y = 'a';
y++;//Valid = 'b'

boolean y = true;
y++;//Invalid. C.E: Operator ++ cannot be applied boolean

~~~

#### Differnce between b++ and b=b+1

##### Ex: 1
If we apply any arithemetic between two variables a and b the result type is always max(int, type of a, type of b)
~~~java
byte x = 10;
byte y = 10;

byte c = x+y;//Invalid

C.E = Possible loss of precision, req int found byte

byte c = (byte)(x+y);//Valid
~~~

##### Ex: 2
~~~java
byte b = 10;
b = b+1;//max(int, byte, int)
//C.E: P.L.P: Found int, req: byte

b = (byte)(b+1)//Valid
    
//But in the case of In/Decrement operators, internal typecasting happens automatically.
    
b++;//Valid
~~~

### Infinity

In integral arithemetic byte, short, int, long there is not way to represent infinity, hence if infinity is result, we will get arithemetic exception in integral arithemetic.

System.out.println(10/0);RE: Division by zero

But in float and double there is a way to represent Infinity for this Float and Double classes contains the following two constants. POSITIVE_INFINITY, NEGATIVE_INFINITY. Hence, even though result is infinity we would not get any error using float, double.

Sop(10/0.0);//Infinity
Sop(-10.0/0.0);//-Infinity

### Undefined NaN - Not a Number
In intergral arithemetic byte, short, int, long, there is no way to define undefined results, hence we get R.E: Arithemetic Exception.
Sop(0/0);//Error

But in floating point arithemetic, there is a way to represent undefined results. Float and Double contains NaN constant.
Hence, if the result is undefined we won't get any Arithemetic Exception.

Sop(0.0/0);//NaN
Sop(-0/0.0);//NaN

#### Note: 
For any x values including NaN the following results in **false**

x<Float.NaN

x<=Float.NaN

x>Float.NaN

x>=Float.NaN

x==Float.NaN

For any x value including NaN the following results in **true**

x!=Float.NaN
