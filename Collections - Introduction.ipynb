### Introduction

An arrray is an indexed collection of fixed number of homogenous data elements. The main advantage of arrys is: we can represent multiple values by using single variable. so that readability of code is improved.

### Limitation of Arrays

1. Arrays are fixed in size. once we create an array, there is not chance of increasing or decreasing size.Due to this we should know the size always, which is not possible always.

2. Array can hold only homogenous data type elements.

~~~java
Student []s = new Strudent[100];
s[0] = new Student();
s[1] = new Customer();//C.E: incompatable type, found Cus, req: Student

// We can solve this by using Object array

Object[] o = new Object[1000];
o[0] = new Student();
0[1] = new Customer();
~~~

3. Arrays concept is not implemented based on some standard data structure, hence ready made method support is not avaiblable. For every req we have to write the code explicitly which increases the complexity of programming.


To over come above problems of arrays we should go for **Collections** concept.

## Collections

1. Collections are growable in nature, i.e., based on our req we can increase or decrease the size.

2. Collections can hold both homogenous and hetrogenous objects.

3. Every collection class is implemented based on some standard data structure, hence for every req ready made support is available, being a programmer we are responsible to use those methods and we are not responsible to implement those methods.

###  Collection(I)
1. If we want to represent a group of individual obj as a single entity then we should go for collection.

2. Collection Interface defines the most common methods, we are applicable for any Collection object.

3. In general, collection interface is considered as root interface of collection framework.

There is no concerete class which implements collection interface directly.

### Queue(I)

It is the child interface of Collection, if we want to represent a group of individual objects prior to processing, then we should go for queue. Usually Queue follows FIFO, but based on our req we can implement our own priority order also. 

Ex: Before sending a mail, all mail ids we have to store in some data structure. In which order we added mail ids in the same order only mail should be delived for this req Queue is the best choice.

### Note:
All the above interface, meant for representing a group of individual objects. If we want to represent a group of objs as key-value pairs then we should go for Map.

### Map(I)

Map is **not** child interface of Collection, if we want to rep a group of objs as key: value pair then we should go for map.

Both key, value are objects, duplicate keys are not allowed but values can be duplicated.
