### Data Types

1. In Java every variable and exp has some type, each and every data type is clearly defined.
2. Every assignment is checked for type compatability. Java is a strongly typed programming language.
int i = 0.5; // invalid
3. Java is not considered as a pure object oriented programming language, several oops features are not satisfied by java, for e.g., operator overloading, multiple inheritance.

### Integral - byte, short, int, long

#### byte

Size: 1 byte - 8 bitss

First bit (MSB) - Acts as signal bit 0 means positive, 1 means negative. 

MAX_VAL = 127
MIN_VAL = -128
Range: -128 to 127

~~~java
byte i = 10; //Valid
byte i = 10.5;//Invalid C.E: P.L.P: found double req: byte

byte i = 128; //Invalid C.E: P.L.P: found int req: byte

byte i = true; //Invalid C.E: incompatible types
~~~
When to use byte: Byte is the best choice if we want to handle data in streams or read from file or network.
File supported and network supported are byte.

#### short

Size: 2 bytes - 16 bits
1 bit for MSB
MAX_VAL = 2^15 - 1
MIN_VAL = -2^15
Range: -32768 to 32767

~~~java
short s = 32786; //Invalid C.E: P.L.P: found int req: short
short s = 10.5;//Invalid C.E: P.L.P: found double req: short
short s = "rajath"; //Invalid C.E: incompatible types
~~~
This was created as 8085/8086 processors mainly used 16 bit.

#### int

Size: 4 bytes - 32 bits
1 bit for MSB
MAX_VAL = 2^31 - 1
MIN_VAL = -2^31
Range: -2147483648 to 2147483647

~~~java
int x = 2147483648; //Invalid C.E: Interger to large
int x = 2147483648l; //Invalid C.E: P.L.P: found double req: int
~~~

#### long

Used for dealing with large numbers, of reading of number of character in a file.

Size: 8 bytes - 64 bits
1 bit for MSB
MAX_VAL = 2^63 - 1
MIN_VAL = -2^63
Range: -2147483648 to 2147483647

1. Old languages like C/C++ char is represented in 1 byte, because it supported only ASCIII values.

Size: 0 - 2^15-1, 0-65535
