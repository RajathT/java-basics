### instanceof

We can use instanceof operator to check whether the given object is of a particular type of not.

~~~java
Object o = l.get(0);

if (o instanceof Student){
    Student s = (Student)o;
}
~~~

Syntax: r instanceof x

r - object reference
x - class or interface

Ex 1:
~~~java
Thread t = new Thread();

System.out.println(t instanceof Thread);
System.out.println(t instanceof Object);
System.out.println(t instanceof Runnable);

//t is a child of Obj, implements Runnable
~~~

Ex 2:
To use instance of operator, there should be some relation between arg types either Parent to child or child to parent or same type, else we get C.E: inconvertable type, found : thread, req String
~~~java
Thread t = new Thread();

System.out.println(t instanceof String);//C.E
~~~

Note: for any class or interface x null instanceof x is always false.

System.out.println(null instanceof x);//false


### Bitwise operators(&, |, ^)

#### & => AND - Returns true iff both args are true
#### | => OR - Returns true iff atleast one args are true
#### ^ => X-OR - Returns true iff both args are different

We can apply these operators for integral types also.
~~~java
System.out.println(4&5);//4
System.out.println(4|5);//5
System.out.println(4^5);//1
~~~

### Bitwise complement operator(~)
We can apply this operator only for integrals type but not for boolean type. If we are trying to apply for boolean type, then we get compile time error. 

System.out.println(~4);//-5

It is not 3 because we should not apply the negation for only the three bits, we apply it to 4 bytes as int is 4 bytes.

32 bits of 4 = 00000....0100
32 bits of ~4 = 11111....1011
As the MSB is 1 = negative, we have to perform 2s complement for the remaining numbers:
1. perform 1s complement
2. add 1
32 bits of ~4 = 10000...0100(1s complement)
32 bits of ~4 = 10000...0101(add 1)

#### Note: 
MSB acts as signed bit, 0 - postive 1 means - negative, positive number will be represent directly, negative will be represent indirectly in 2s complement form.
