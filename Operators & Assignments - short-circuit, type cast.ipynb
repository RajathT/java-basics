#### short circuit operator

These are similar to bitwise, except the following differences:

##### &, | - 
1. Both args should be evaluated always
2. Performance is low
3. Applicable for both boolean and for intgreal types
##### &&, || - 
1. Second args evaluation is optional
2. Performance is high
3. Applicable for only boolean but not intgreal types

1. x&y = y will be evaluated iff x is true
2. x||y = y will be evaluated iff x is false

##### Ex: 1
~~~java
int x = 10, y=15;
if (++x < 10 & ++y<15){
    x++;
}
else{
    y++;
}
~~~

|double 11|16|
|single or 12|16|
|double 12|16|

#### Type cast operator

There are two types of typecasting

1. Implicit typecasting:
    * Compiler is responsible to perform implicit typcasting
    * Whenever we are assigning smaller DT val to bigger DT val, implicit typecasting
    * This is called widening or upcasting
    * No loss of information
int x = 'a';
Sop(x); //97

double d = 10;
Sop(d); //10.0

2. Explicit typecasting
    * Programmer is responsible to perform this
    * Whenever we are assigning bigger DT to smaller DT
    * It is also known as narrowing or downcasting
    * There maybe a chance of loss of information
    
##### Ex 1:
int x = 130;
byte b = x;
S.o.p(b); C.E: possible loss of information

int x = 130;
byte b = (byte)x;
S.o.p(b);//-126 

Whenever we are assigning bigger DT val to smaller DT val by explicit typecasting the MSB bits will be most, we have to consider LSB.

130 = 0000.....010000010
as it is converted to byte only last 8 are stored
