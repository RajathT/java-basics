#### Coding Standards for Classes

Usually class names are nouns should start with upper case letter, and if it has multiple words then it should have uppercase letter. StringBuffer

#### Coding Standards for Interface

Usually interface names are adjective, should start with uppercase letters and it has multiple words then it should have uppercase letter. Runnable, Serializable, Comparable

#### Coding Standards for Methods

Usually method names are verbs or verb-noun combination, should start with lower, if contains multiple words, inner words are upper case. print(), getName(), getSalary(). CamelCase convention.


#### Coding standards for variables

Usually variable names are nouns, should start with lowercase letters and it has multiple words then it should have uppercase letter. name, age, mobileNumber

#### Coding standards for constants

Usually contant names are nouns, should contain only uppercase characters and if it contains multiple words, then these words are seperated with underscore. MAX_VALUE, MIN_PRIORITY.

Note: Usually we can declare constants with, public, static and final modifiers.


#### Coding standards for Java Bean

Java Bean - Is a simple Java class with privatei variables are private and public getter and setters for every variable.

~~~java
public class StudentBean{
    private String name;
    
    public void setName(String name){
        this.name = name;
    }
    public void getName(){
        return name;
    }
}
~~~

#### Syntax for setter method

1. It should be public
2. return type should be void
3. Method name should be prefixed with set, setName
4. It should take some argument

#### Syntax for getter method

1. It should be public
2. return type should not void
3. Method name should be prefixed with get, getName
4. It should take no argument

* For boolen properties, getter name can be prefixed with get or is, but recommeneded to use is, isEmpty(), getEmpty()

### Coding standards for Listeners

public void addMyActionListner(MyActionListner l);
