### Identifiers
A name in Java program is called identifier, which can be used for identification purpose, it can be method name, variable name, class name or label name.

~~~java
class Test{
    public static void main(String []args){
        int x = 10;
    }
}
~~~
Test, main, String, args, x all are itdentifiers

1. Valid Identifier rules: a-z, A-Z, $, _ any other you get C.E

2. total123 is correct, 123total is incorrect
3. Java is case sensitive: number, Number all are different
4. No limit on the length of identifier
5. Cannot use reserved words as identifiers.
int x = 10; //Valid
int if = 10; //Invalid

6. All predefined class names, interface names, can be used as identifiers, but not a good coding practice.
int String = 888;
System.out.println(String);//Valid: O/P 888

int Runnable = 888;
System.out.println(Runnable);//Valid: O/P 888
