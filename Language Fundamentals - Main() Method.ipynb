Whether class contains main() or not whether main() method is declared according to main method or not, these things will not be checked by the compiler, at run time JVM will check these. If JVM doesn't find the main() then R.E: NoSuchMethodError: main

~~~java
class Test{
}
// javac Test.java No error
// java Test Error
~~~

At runtime JVM always searches for the main method with the following prototype.
#### public - To call by JVM from anywhere
#### static - Without existing obj also, JVM has to call this method
#### void - main method won't return anything to JVM
#### main - this is configured inside JVM so it has to be main
#### String[]args - Command line args

The above syntax is very strict is we perform any change we will get R.E : NoSuchMethodError: main

Even though the above syntax is very strict the following changes are acceptable.
* Instead of public static we can take static public i.e., the order of modifiers is not important. We can declare string [] in any acceptable form main(String args[]), main(String []args)

* Instead of args we can take any valid java identifier main(String[]rajath)

* we can replace (String[]args) with (String...args)

* we can declare the main method with the following modifiers. final, synchronized, strictfp
~~~java
class Test{
    static final synchronized strictfp public void main(String..args){
        System.out.println("Valid");
    }
}
~~~

Which of the following main method are valid:
~~~java
public static void main(String args);//Invalid
public static void Main(String[]args);//Invalid
public void main(String []args);//Invalid
public static int main(String []args)//Invalid
final synchronized strictfp public void main(String[] args);//Invalid
final synchronized strictfp public static void main(String[] args);//Valid
public static void main(String..args);//Valid
~~~

Which of the above we get compile time errror

No wont get compile time error anywhere except last two cases we get R.E : NoSuchMethodError: main

### Case 1:

Overloading of main method is possible, but JVM will always call String[]args method only, the other overloaded method we have to call explicitly, like normal method call.

~~~java

class Test{
    public static void main(String[]args){
        System.out.println("String[]");
    }
    public static void main(int[]args){
        System.out.println("int[]");
    }
    //Valid but JVM will call String[]
}
~~~

### Case 2:
Inheritance concept applicable for main method, hence while executing child class if a child doesn't contain main, then parent class main will be executed.
~~~java
class P{
    public static void main(String[]args){
        System.out.println("Parent");
    }
}
class C extends P{
}

~~~
javac P.java

This will give two class files P and C

java P -> Output: Parent

java C -> Output: Parent

### Case 3:

~~~java
class P{
    public static void main(String[]args){
        System.out.println("Parent");
    }
}
class C extends P{
    public static void main(String[]args){
        System.out.println("Child");
    }
}

~~~
javac P.java
This will give two class files P and C

java P -> Output: Parent

java C -> Output: Child

This is example of method hiding not overriding.

It seems overrding concept applicable for main method but it is not overriding but method hiding.

For main method Inheritance and overloading concepts are applicable but overriding concept is not applicable, instead of overriding method hiding is applicable.
