### String Concatenation Operator
The only overloaded operator in Java is the + operator, sometimes it acts as arithemetic addition operator and sometimes it acts as a string concatenation operator.

If atleast one argument is String type then + acts as concatenation, and if both are Number type the + acts as arithemetic addition operator.

#### Case :1
~~~java
String a = "rajath";
int b=10, c=20,d=30;

System.out.println(a+b+c+d);//rajath102030
System.out.println(b+c+d+a);//60rajath
System.out.println(b+c+a+d);//30rajath30
System.out.println(b+a+c+d);//10rajath2030
~~~
#### Case :2
~~~java
String a = "rajath";
int b=10,c=20,d=30;

a=b+c+d;//C.E: Incompatible types
a=a+b+c;
b=a+c+d;//C.E: Incompatible types
b=b+c+d;
~~~


### Relational Operators (<,<=,>,>=)

#### Case :1
We can apply relational operators for every primitive type except boolean.
~~~java
System.out.println(10<20);
System.out.println('a'<20);
System.out.println('a'<97.8);
System.out.println('a'<'A');
System.out.println(true<false);//C.E Operator > cannot be applied to boolean, boolean.
~~~

#### Case: 2
We can't apply relational operators for object types
~~~java
System.out.println("rajath123">"rajath");//C.E: operator > cannot be applied to java.lang.String, java.lang.String
~~~

#### Case: 3

Nesting of relational operators is not allowed otherwise we will get error.

~~~java
System.out.println(10<20<30);//C.E: operator > cannot be applied to boolean, int
~~~



### Equality operators (==, !=)

We can apply equality operators for every primitive type including boolean type.

#### Case 1:
~~~java
System.out.println(10==20);//false
System.out.println('a'=='b');//false
System.out.println('a'==97.8);//true
System.out.println(false==false);//true
~~~

#### Case 2:

We can apply equality operators for obj also, for obj ref r1, r2
r1 == r2 return true if and only if both ref pointing to a same obj. Ref comparision or address comparision.

    r2-/
    
##### Ex1:
~~~java
Thread t1 = new Thread();
Thread t2 = new Thread();
Thread t3 = t1;

System.out.println(t1==t2);//false
System.out.println(t1==t3);//true
~~~

##### Ex2:

~~~java

Object o = new Object();
~~~

### Diff between == operator and equals()

In general we can use == operator for ref comparision and equal() for content comparision. 

~~~java
String s1 = new String("rajath");
String s2 = new String("rajath");

System.out.println(s1==s2);//false
System.out.println(s1.equals(s2));//true
~~~

##### Note:
For any obj ref r, r==null is always false. but null == null is true.

~~~java
String s = new String("rajath");
System.out.println(s==null);//false

String s = null;
System.out.println(s==null);//true

System.out.println(null==null);//true
~~~
