### Assignment operators
There are 3 types:
1. Simple assignments: int x = 10;
2. Chained assignment: 
a=b=c=d=10;
we can perform chained assignment directly at the time declaration.
int a=b=c=10;//C.E we cannot find symbol: variable b location: class test
int b,c;
int a=b=c=20;//Valid

3. Compund assignment operator
Sometime assignment operator mixed with other operators, such type of assignment operators are called compund operators.

int a = 20;
a += 10;
S.o.p(a);//30

All possible compound assignments:

+=, -=, *=, /=, %=, &=, |=, ^=, >>=, >>>=, <<=. No <<<=.

In the case of compound operators internal typcasting will happen automatically.

byte b = 10;
b = b+1; //max(int, byte, int)
S.o.p(b);
//C.E: P.L.P: found int, req: byte

byte b = 10;
b++;//b=(byte)(b+1);
S.o.p(b);//11 

byte b = 10;
b+=1;//b=(byte)(b+1);
S.o.p(b);//11

Conditional operator(?:)

The only possible ternary operator in java is conditional operator.

int x = (10<20)?30:40
S.o.p = 30

We can perform nesting of conditional operator also

int x = (10>20)?30:(40>50?60:70)

S.o.p x = 70

#### new operator
We can use new operator to create obj.

For e.g. Test t = new Test();

After creating an object constructor will be executed to perform for initialization of obj, hence constructor is not for creation of obj but for initialization of the obj.

We have new but not delete, destruction of useless obj is responsibilty of garbage collector.
