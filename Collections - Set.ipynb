### HashSet
1. The under lying datastructure is hash tale
2. Duplicate elements are not allowed
3. Insertion order is not preserved, and it is based on Hash code of objects.
4. null insertion is possible, only once.
5. Hetrogenous values are allowed
6. Implements Serializable, Cloneable but not RandomAccess Interface.
7. HashSet is best choice if we have search operations.

#### Note: 
Duplicates are not allowed, if we are tyring to insert duplicates then we won't get any comiple time or run time errors, simply return false

~~~java

HashSet h = new HashSet();
h.add("A");//true
h.add("A");//false
~~~

#### Constructors
~~~java
HashSet h = new HashSet();
//Create an empty HashSet with default value of 16, and default fill ratio 0.75.

HashSet h = new HashSet(int initialcapacity);
//Create an empty HashSet object with specified intitial capacity and a default ratio 0.75.

HashSet h = new HashSet(int initialcapacity, float fillRatio);

HashSet h = new HashSet(Collection c);
//Create an equivalent HashSet for the given Collection., meant for inter converstion between collection objects.

Fill Ratio/Load Factor: After filling 0.75, create new HashSet Object. 
~~~

### SortedSet

SortedSet is the child interface of Set, if we want to represent a group of individual objects, according to some sorting order without duplicates, then we should go for sorted set.

### TreeSet
1. The underlying DS is balanced Tree
2. Duplicate objects are not allowed
3. Insertion order not preserved, default sorting.
4. Hetrogenous objects are not allowed, otherwise R.E: ClassCastException.
5. null insertion not available.
6. TreeSet implements Serializable, Cloneable but not RandomAccess
7. All objects will be inserted based on some sorting order it maybe default, natural order or customized sorting order

### Constructors

~~~java
TreeSet t = new TreeSet();
//creates an empty TreeSort, where the elements will be inserted to natural sorting order

TreeSet t = new TreeSet(Comparator c);
//create an empty TreeSet obj, where the elements will be inserted by customized sorting order using comparator obj

TreeSet t = new TreeSet(Collection c);

TreeSet t = new TreeSet(SortedSet s);
//Will use SortedSet sorting order to sort
~~~

~~~java
TreeSet t = new TreeSet();
t.add("A");
t.add("a");
t.add("B");
t.add("Z");
System.out.println(t);//[A, B, Z, a]
~~~


For non-empty TreeSet if we are trying to insert null, we will get NPE.
For empty TreeSet the first element is allowed, but after inserting that null, if we are trying to insert any other then we will get runtime exception saying null pointer exception.
 
Until 1.6v null is allowed as the first element to the empty TreeSet, but from 1.7v null is not allowed.

~~~java
TreeSet t = new TreeSet();
t.add(new StringBuffer("A"));
t.add(new StringBuffer("B"));
t.add(new StringBuffer("C"));
t.add(new StringBuffer("D"));
System.out.println(t);//R.E: ClassCastException
~~~
If we are depending on default natural sorting order then they should be homogenous and comparable. else we will get R.E: ClassCastException.

An object is said to be comparable iff corresponding class implements Comparable interface, String Class and all Wrapper Class implement Comparable Interface, but StringBuffer class doesn't implement Comparable interface.
Hence we got RE: CCE in the above example.

### Comparator(I)
Present in java.util package and defines two methods:
public int compare(Object obj1, Object obj2);

|Check|Return Value|
|-|-|
|iff obj1 has to come before obj2| -ve|
|iff obj2 has to come before obj1| +ve|
|obj1 and obj2 are equal|0|

public boolean equals(Object obj);

Whenever we are implementing Comparator interface, we provide implementation for only Compare method, and we are not req to provide implementation for equals method because it is already available to our class from Object class through inheritance.

Write a program to insert integer objects to TreeSet where sorting order is descending order:

```scijava
import java.util.*;
class TreeSetDemo{
    public static void main(String[] args){
        TreeSet t = new TreeSet(new myComparator());
        for(int i = 0; i<10; i++)
            t.add(i);
        System.out.println(t);
    }
}
class myComparator implements Comparator{
    public int compare(Object obj1, Object obj2){
        Integer i = (Integer)obj1, j = (Integer)obj2;
        
        if (i>j)
            return -1;
        else if (i<j)
            return 1;
        else
            return 0;
        
        /*
        
        Various possible implementations of Compare method:
        
        Method: To gives ascending:
        return i.compareTo(j);
        */
        
        /*
        Method: To gives decending, negative of what compareTo give
        return - i.compareTo(j);
        */
        
        /*
        Method: To gives decending, reverse the values of compareTo
        return j.compareTo(i);
        */
        
        /*
        Method: To gives ascending, reverse two times
        return -j.compareTo(i);
        */
        
        /*
        Keep inserting everthing after obj1
        returns : order in which it was inserted
        return 1;
        */
        
        /*
        Keep inserting everthing before obj1
        returns : reverse of the order in which it was inserted
        return -1;
        */
        
        /*
        After first element everything is duplicate
        return 0;
        */
     
    }
}
```

At line 1 if we are not passing comparator, then internally JVM will call compareTo method which is meant for default sorting order, in this case = 0,1,...9.
At line 1 if we pass the Comparator method, then JVM will call compare method which is meant for customized sorting.
We get: [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]

### Comparable vs Comparator

1. For predefined Comparable classes default natural sorting order already available, if we are not satisfied with that defualt sorting order, then we can define our own sorting by using comparator.

2. For predefined non-Comparable classes like StringBuffer, default natural sorting not available, we can define our own sorting by using Comparator.

3. For our own classes like Employee, the person who is writing the class is responsible to define default natural sorting order by implementing Comparable interface. The person who is using, if he is not satisfied, then he can define his own sorting using Comparator.

```scijava
import java.util.*;

class Employee implements Comparable{
    
    String name;
    int id;
    
    Employee(){}
    
    Employee(String name, int id){
        this.name = name;
        this.id = id;
    }
    
    public String toString(){
        return id + " " + name;
    }
    
    public int compareTo(Object obj){
        int val_1 = this.id;
        Employee e = (Employee)obj;
        int val_2 = e.id;
        
        if (val_1 < val_2)
            return -1;
        else if(val_2 < val_1)
            return 1;
        else
            return 0;
    }
    
    public static void main(String[]args){
        System.out.println("Hello");
        TreeSet s = new TreeSet(new MyComparator());
        s.add(new Employee("rajath",1));
        s.add(new Employee("alok",2));
        s.add(new Employee("vidya",3));
        System.out.println(s);
    }
}

class MyComparator implements Comparator{
    public int compare(Object obj1, Object obj2){
        Employee e1 = (Employee)obj1;
        Employee e2 = (Employee)obj2;
        return (e1.name).compareTo(e2.name);
    }
}
```

### Comparable vs Comparator

|defines only one method compareTo()|defines compare() and equals()|
|String and all wrapper classes implement Comparable|Only implement classes of Comparator are Collator, RuledBasedCollator|

### NavigableSet(I)

As the part of 1.6v following introduced in Collections Framework

1. It is the child interface of SortedSet, and defines several methods for navigation purposes.

### Methods
~~~java
floor(e)//returns highest element <= e
lower(e)//return highest element which is <e
ceiling(e)//returns lowest element which is >= e
higher(e)//returns lowest element which is > e
pollFirst()//remove and return first element
pollLast()//remove and return last element
desendingSet()//returns NavigableSet in reverse order
~~~

```scijava
class NavigableSetDemo{
    public static void main(String[] args){
        TreeSet<Integer> t = new TreeSet<Integer>();
        t.add(1000);
        t.add(2000);
        t.add(3000);
        t.add(4000);
        System.out.println(t);
        System.out.println(t.ceiling(2000));
        System.out.println(t.higher(2000));
        System.out.println(t.floor(3000));
        System.out.println(t.lower(3000));
        System.out.println(t.pollFirst());
        System.out.println(t.pollLast());
        System.out.println(t.descendingSet());
        System.out.println(t);
    }
}
```
